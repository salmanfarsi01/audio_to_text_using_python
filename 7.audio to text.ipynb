```python
import speech_recognition as sr
from googletrans import Translator
```

```python
def main():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        r.adjust_for_ambient_noise(source)
        print("Please Say Something....")
        audio = r.listen(source)
        try:
            print("You have said:\n" + r.recognize_google(audio))
        except Exception as e:
            print("Error:" +str (e))

if __name__ == "__main__":
    main()
```

```python
from docx import Document

def main():
    r = sr.Recognizer()
    
    # Path to your audio file (replace with the path to your file)
    audio_file_path = "C:\\Users\\Local User\\Desktop\\python_2308\\voice data\\record1.wav"
    
    with sr.AudioFile(audio_file_path) as source:
        audio = r.record(source)  # Record the audio from the file
        
        try:
            transcription = r.recognize_google(audio)
            print("Transcription:\n" + transcription)
            
            # Create a new Document
            doc = Document()
            doc.add_heading('Transcription', 0)
            
            # Add the transcription to the document
            doc.add_paragraph(transcription)
            
            # Save the document
            doc.save("transcription.docx")
            print("Transcription saved as 'transcription.docx'")
        
        except Exception as e:
            print("Error:" + str(e))

if __name__ == "__main__":
    main()
```
